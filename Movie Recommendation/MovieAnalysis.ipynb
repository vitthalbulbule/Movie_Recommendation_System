```python
# We want required to convert or access only genres Name
import ast

def convert(obj):
    if obj is None or obj == '' or str(obj) == 'nan':
        return []   # return empty list if missing

    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```
